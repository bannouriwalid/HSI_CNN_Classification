<h3 style="color: #0077b6; font-family: 'Times New Roman'; font-weight: bold;">
    a. Définition des images hyperspectrales
</h3> 

<div style="font-family: Garamond; font-size: 18px;">
    
Les images hyperspectrales (HSI) sont des images acquises par des capteurs capables de mesurer **la réflectance** du sol ou d’un objet sur un grand nombre de bandes spectrales **étroites** et **contiguës**, souvent réparties dans les domaines du visible, du **proche infrarouge (NIR)**, voire de **l’infrarouge moyen (MIR)**.
    
Contrairement aux images traditionnelles (comme les images RGB), qui comportent uniquement trois canaux (rouge, vert, bleu), les images hyperspectrales peuvent contenir **des centaines bandes**, chacune correspondant à une **longueur d’onde spécifique**. Chaque pixel de l’image est donc associé à un **vecteur spectral** de grande dimension, représentant sa **signature spectrale unique**.

Cela confère aux images hyperspectrales **une capacité exceptionnelle à discriminer les matériaux** en fonction de leur **réponse spectrale**, ce qui est très utile dans des domaines comme l'agriculture, la géologie ou la détection de pollution.
</div>

<h3 style="color: #0077b6; font-family: 'Times New Roman'; font-weight: bold;">
    b. Particularités des images hyperspectrales
</h3>

<div style="font-family: Garamond; font-size: 18px;">

Les images hyperspectrales se distinguent par plusieurs caractéristiques techniques majeures :

* **Haute dimensionnalité :**
Chaque pixel est représenté par un vecteur de plusieurs centaines de dimensions. L’image peut donc être vue comme **un cube 3D** (lignes, colonnes, bandes spectrales).
* **Redondance spectrale :**
Les bandes étant très proches les unes des autres, il existe **une forte corrélation** entre elles. Certaines bandes peuvent même être **redondantes** ou **bruitées**, ce qui peut dégrader les performances des algorithmes. Pour y remédier, on utilise des techniques de réduction de dimensionnalité, telles que **l’Analyse en Composantes Principales (PCA)** ou t-SNE, afin de compresser l’information tout en conservant les caractéristiques discriminantes.
* **Volume de données élevé :**
Le nombre important de bandes et de pixels rend chaque image volumineuse (plusieurs centaines de Mo). Cela impose des contraintes de traitement, de stockage, et d’apprentissage, notamment pour les modèles profonds.

</div>

<h3 style="color: #0077b6; font-family: 'Times New Roman'; font-weight: bold;">
    c. Problématique de la classification d’images hyperspectrales
</h3>

<div style="font-family: Garamond; font-size: 18px;">

L’objectif principal de la classification hyperspectrale est d’assigner à chaque pixel une étiquette correspondant à un type de matériau ou d’objet présent dans la scène. Toutefois, plusieurs défis majeurs rendent cette tâche complexe :
* **Curse of dimensionality :** l’augmentation du nombre de dimensions augmente la complexité des modèles et le risque de sur-apprentissage, surtout avec peu de données étiquetées.
* **Imbalance des classes :** certaines classes sont surreprésentées tandis que d’autres sont rares.
* **Mixing spectral :** plusieurs matériaux peuvent coexister dans un même pixel (phénomène de mélange).
* **Besoin de généralisation :** un modèle doit être capable de bien se comporter sur des scènes différentes, avec des conditions de lumière et des capteurs variés.
</div>

<h2 style="color: #023e8a; font-family: 'Times New Roman'; font-weight: bold;">
    3. Réduction de dimension : PCA
</h2>

<div style="font-family: Garamond; font-size: 18px;">

Les images hyperspectrales se caractérisent par une très **haute dimensionnalité** et d'une **redondance spectrale** (de nombreuses bandes sont fortement corrélées). Cela pose plusieurs défis en classification avec des réseaux convolutifs (CNN) comme **la surcharge computationnelle** et **l'augmentation du risque d’overfitting**.

ACP (Analyse en Composantes Principales) est une technique couramment utilisée pour réduire la dimensionnalité tout en conservant l’essentiel de la variance. Elle permet de :
* **Projeter les données dans un espace de plus faible dimension**, plus adapté à l’apprentissage.
* **Conserver l'information spectrale dominante** dans les premières composantes.
* Accélérer l’entraînement des CNN tout en réduisant la complexité du modèle.

Il est essentiel de **centrer les données** avant d'appliquer le PCA, c’est-à-dire soustraire la moyenne de chaque bande spectrale. Cela garantit que le PCA capte la variance réelle dans les directions principales, non biaisée par des décalages moyens.
</div>

    III. Classification des images hyperspectrales par CNN
</h1>

<div style="font-family: Garamond; font-size: 18px;">

Les réseaux de neurones convolutifs, grâce à leur capacité à extraire automatiquement des caractéristiques hiérarchiques et à modéliser des relations complexes, se sont imposés comme une solution efficace pour la classification de ces données riches, à condition d’adapter leur architecture aux spécificités des HSI.

<h2 style="color: #023e8a; font-family: 'Times New Roman'; font-weight: bold;">
    1. Description des différentes étapes à suivre
</h2>

<h3 style="color: #0077b6; font-family: 'Times New Roman'; font-weight: bold;">
    a. Architecture adoptée
</h3>

<div style="font-family: Garamond; font-size: 18px;">

Le modèle conçu repose sur une architecture **hybride combinant des convolutions 3D et 2D**, afin d’exploiter simultanément les informations spectrales (bandes hyperspectrales) et spatiales (voisinage des pixels) contenues dans les patches extraits de l’image.

Voici le détail de l’architecture :

* **Entrée :** L’entrée du réseau est un tenseur de forme (S, S, L, 1), où S est la taille spatiale du patch (par exemple 25×25), L le nombre de composantes spectrales (réduites via PCA), et 1 représente une seule "pseudo-canal" pour la compatibilité avec les couches 3D.

* **Bloc de convolutions 3D :**
Trois couches Conv3D successives sont utilisées pour extraire des caractéristiques conjointes spatiales et spectrales :

    * Conv3D(8 filtres, 3x3x7)
    * Conv3D(16 filtres, 3x3x5)
    * Conv3D(32 filtres, 3x3x3)
* **Passage des convolutions 3D à 2D :** Après les convolutions 3D, le tenseur est remodelé avec Reshape pour fusionner la dimension spectrale restante avec la profondeur des canaux, donnant un tenseur de forme (H, W, C) compatible avec les convolutions 2D.

* **Convolution 2D :** Une couche Conv2D avec 64 filtres de taille (3x3) est appliquée pour raffiner davantage les caractéristiques spatiales.

* **Couches entièrement connectées (denses) :**
    * aplatir les cartes de caractéristiques en un vecteur.
    * Deux couches Dense successives avec 256 et 128 neurones (activation ReLU), chacune suivie d’un Dropout(0.4) pour réduire le surapprentissage.
    * Une couche de sortie Dense avec une activation softmax, adaptée à un problème de classification multi-classes, produit les probabilités finales.
</div>

<h3 style="color: #0077b6; font-family: 'Times New Roman'; font-weight: bold;">
    b. Prétraitement des données hyperspectrales
</h3>

<div style="font-family: Garamond; font-size: 18px;">

Le prétraitement des données joue un rôle central dans l’exploitation efficace des images hyperspectrales, notamment lorsqu’on les utilise comme entrée pour un réseau de neurones convolutionnel. Le pipeline adopté dans notre projet suit les étapes clés suivantes :  
**1. Réduction de dimension par ACP (PCA):**
Afin de conserver uniquement l'information la plus pertinente, nous appliquons une Analyse en Composantes Principales (ACP) pour réduire la dimension spectrale à un nombre fixe de composantes principales (fixé précedemment), tout en conservant la majorité de la variance.  
**2. Extraction de patches autour de chaque pixel:**
Pour chaque pixel de l’image, nous extrayons une petite fenêtre (ou patch) centrée sur celui-ci, de taille fixe (25x25 par exemple). Chaque patch capture à la fois l'information spectrale (les composantes réduites) et spatiale (le voisinage du pixel), ce qui est essentiel pour permettre au modèle de tirer parti des deux dimensions.  
**3. Division des données en ensembles d’entraînement et de test:**
Une fois les patches extraits, nous répartissons les données en ensembles d’apprentissage et de test. Cette séparation est réalisée à l’aide de train_test_split, en stratifiant les échantillons sur les classes (stratify=y) afin de préserver la distribution des classes dans chaque sous-ensemble.  
**4. Mise en forme pour le réseau convolutionnel 3D:**
Les données sont ensuite remodelées afin d’être compatibles avec l’entrée du réseau 3D, sous la forme (nombre_exemples, hauteur_patch, largeur_patch, composantes, 1). Les labels sont convertis au format one-hot encoded pour l’apprentissage multi-classes.
</div>

<h3 style="color: #0077b6; font-family: 'Times New Roman'; font-weight: bold;">
    c. Création et compilation du modéle
</h3>

<div style="font-family: Garamond; font-size: 18px;">

Une fois le modèle construit, nous procédons à sa compilation en définissant les paramètres clés suivants :

* **Optimiseur : Adam** (utilisé par défaut), reconnu pour sa rapidité de convergence et sa capacité d’adaptation grâce à l’ajustement dynamique du taux d’apprentissage.
* **Fonction de perte : categorical_crossentropy**, particulièrement adaptée aux tâches de classification multi-classes, où les étiquettes sont encodées en one-hot.
* **Métrique d’évaluation : accuracy**, permettant de suivre la performance du modèle aussi bien sur les données d’entraînement que de validation.

Le modèle est ensuite entraîné sur les données prétraitées via un nombre défini d’époques, avec une taille de batch fixe.
</div>

<h3 style="color: #0077b6; font-family: 'Times New Roman'; font-weight: bold;">
    d. Evaluation du modèle
</h3>

<div style="font-family: Garamond; font-size: 18px;">

l'évaluation de la performance et la robustesse du modèle repose sur les métriques suivantes:
* **Overall Accuracy (OA) :** proportion globale de bonnes prédictions.
* **Average Accuracy (AA) :** moyenne des taux de bonnes prédictions pour chaque classe.
* **Coefficient de Kappa (κ) :** mesure de l’accord entre prédictions et vérité terrain, corrigée du hasard.
</div>
